```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
    'Quantity': [100, 50, 120, 60, 150],
    'Sales': [1000, 500, 1200, 600, 1500],
    'Profit': [200, 100, 250, 120, 300]
}

# Create DataFrame
df = pd.DataFrame(data)

# ✅ Calculate totals
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_quantity = df['Quantity'].sum()

# ✅ Calculate averages per category
average_sales = df.groupby('Category')['Sales'].mean()
average_profit = df.groupby('Category')['Profit'].mean()
average_quantity = df.groupby('Category')['Quantity'].mean()

# ✅ Print results
print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Total Quantity:", total_quantity)

print("\nAverage Sales per Category:\n", average_sales)
print("\nAverage Profit per Category:\n", average_profit)
print("\nAverage Quantity per Category:\n", average_quantity)

# Data Visualization

# Sales & Profit per Category (Bar chart)
category_summary = df.groupby("Category")[["Sales", "Profit"]].sum()

category_summary.plot(kind="bar", figsize=(8,5))
plt.title("Sales & Profit per Category")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(title="Metrics")
plt.show()

# Quantity per Category (Pie chart)
df.groupby("Category")["Quantity"].sum().plot(
    kind="pie", autopct="%1.1f%%", figsize=(6,6), startangle=90
)
plt.title("Quantity Distribution by Category")
plt.ylabel("")  # remove y-label
plt.plot()
plt.show()
```
